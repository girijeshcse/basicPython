**Recreate the swarm plot using titanic dataset, use factorplot** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
```

```python
# Leave this cell empty please.
sns.factorplot(x='class',y='age',data=titanic, col="sex",kind="swarm",palette='rainbow')
```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
```

```python
# Leave this cell empty please.
sns.countplot(x='sex',data=titanic,hue="class")
```

**Recreate the plot below. Use iris dataset. Hint: PairGrid.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
```

```python
# Leave this cell empty please.
g = sns.PairGrid(iris , hue="species")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend();
```

**Recreate the plot below. Use titanic dataset. Hint: FacetGrid.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
```

```python
# Leave this cell empty please.
g = sns.FacetGrid(data=titanic,col='sex', hue='alive')
g.map(plt.hist,'age',alpha=0.5)
g.add_legend()
```

**Recreate the plot below. Use titanic dataset. Hint: Matrix plots** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
```

```python
# Leave this cell empty please.
sns.heatmap(titanic.corr(),cmap='coolwarm',annot=True)
plt.title('titanic.corr()')
```
